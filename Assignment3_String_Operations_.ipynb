```python
# Given string
text = " welcome to python programming "

# Uppercase
print(text.upper())

#Lowercase
print(text.lower())

# Slicing(first 7 characters)
print(text[0:7])

#strip(remove extra spaces)
print(text.strip())

#Concatenation
new_text = text.strip() + "is fine!"
print(new_text)
```
